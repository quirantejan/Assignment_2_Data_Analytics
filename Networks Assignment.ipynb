```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

blue_nodes = ['D', 'F', 'I', 'N', 'S']
green_nodes = ['BIH', 'GEO', 'ISR', 'MNE', 'SRB', 'CHE', 'TUR', 'UKR', 'GBR', 'AUS', 'HKG', 'USA']
yellow_nodes = ['AUT', 'BEL', 'BGR', 'HRV', 'CZE', 'EST', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LUX', 'NLD', 'PRT', 'ROU', 'SVK', 'SVN', 'ESP']

G.add_nodes_from(blue_nodes)
G.add_nodes_from(green_nodes)
G.add_nodes_from(yellow_nodes)

edges = [
    ('D', 'F'), ('D', 'I'), ('D', 'N'), ('D', 'S'),
    ('F', 'I'), ('F', 'N'), ('F', 'S'),
    ('I', 'N'), ('I', 'S'),
    ('N', 'S'), ('D', 'BIH'), ('F', 'GEO'), ('I', 'ISR'), ('N', 'MNE'), ('S', 'SRB'),
    ('D', 'AUT'), ('F', 'BEL'), ('I', 'BGR'), ('N', 'CZE'), ('S', 'EST'),
    ('S', 'AUT'), ('S', 'BEL'), ('F', 'BEL'), ('S', 'BIH'), ('S', 'BGR'),
    ('S', 'HRV'), ('F', 'HRV'), ('S', 'CZE'), ('S', 'EST'), ('S', 'FRA'),
    ('D', 'FRA'), ('F', 'FRA'), ('S', 'GEO'), ('S', 'DEU'), ('N', 'DEU'),
    ('F', 'DEU'), ('F', 'GRC'), ('S', 'GRC'), ('S', 'HUN'), ('S', 'IRL'),
    ('F', 'IRL'), ('S', 'ISR'), ('S', 'ITA'), ('N', 'ITA'), ('F', 'ITA'),
    ('F', 'LVA'), ('S', 'LUX'), ('S', 'MNE'), ('S', 'NLD'), ('N', 'NLD'),
    ('F', 'NLD'), ('S', 'PRT'), ('F', 'PRT'), ('S', 'ROU'), ('F', 'ROU'),
    ('S', 'SRB'), ('F', 'SVK'), ('F', 'SVN'), ('S', 'ESP'), ('N', 'ESP'),
    ('F', 'ESP'), ('D', 'ESP'), ('S', 'CHE'), ('F', 'CHE'), ('S', 'TUR'),
    ('S', 'UKR'), ('S', 'GBR'), ('N', 'GBR'), ('F', 'GBR'), ('D', 'GBR'),
    ('S', 'AUS'), ('N', 'AUS'), ('S', 'HKG'), ('S', 'USA'), ('N', 'USA'),
    ('D', 'USA')
    
]

G.add_edges_from(edges)

color_map = []
for node in G.nodes():
    if node in blue_nodes:
        color_map.append('blue')
    elif node in green_nodes:
        color_map.append('green')
    elif node in yellow_nodes:
        color_map.append('yellow')
    else:
        color_map.append('gray')  

pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_size=500, node_color=color_map, font_size=7, font_color='BLACK', edge_color='black', arrows=True)

plt.title("Network Graph")
plt.show()
```
