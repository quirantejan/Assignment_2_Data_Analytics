```python
import pandas as pd
import matplotlib.pyplot as plt

categories = ['S', 'N', 'I', 'F', 'D']
values_yes = [5, 2, 1, 2, 3]
values_no = [5, 1, 0, 2, 1]

df = pd.DataFrame({
    "Yes": values_yes,
    "No": values_no
}, index=categories)

colors = ['#FF69B4', '#8A2BE2']  

ax = df.plot(kind="barh", stacked=True, figsize=(10, 6), color=colors)

plt.xlabel("Response Count", fontsize=10, fontweight='bold')
plt.ylabel("Labels", fontsize=10, fontweight='bold')
plt.title(
    "Distribution of 'Yes' and 'No' Responses Among Different Categories",
    fontsize=16,  
    fontweight='bold',  
    color='purple'  
)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
